# Links between GDP and Life expectancy

## Introduction
In this project we will investigate the relationship between GDP and Life expectancy for six countries over time.
* Definition GDP: Gross Domestic Product, is one of the primary indicators used to gauge the health of a country's economy. It represents the total dollar value of all goods and services produced over a specific time period within a nation's borders.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#inspect data
df = pd.read_csv('all_data.csv')
df.head()
```

```python
# rename columns Life expectancy at birth (years) for convenience
df.rename({'Life expectancy at birth (years)' : 'LEABY'}, axis=1, inplace=True)
```

```python
#GDP per Country over time
ax = plt.subplot()
plt.title('GDP per Country over time')
sns.lineplot(data=df, x='Year', y='GDP', hue='Country', palette='bright')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.show()

#LEABY per Country over time
ax = plt.subplot()
plt.title('Life expectancy at birth per Country over time')
sns.lineplot(data=df, x='Year', y='LEABY', hue='Country', palette='bright')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.ylabel('Life expectancy at birth (years)')
plt.show()
```

After a brief first look over the data we can say that GDP and Life expectancy rise for each country over time.
To investigate furhter we will plot each country's GDP and Life expectancy over time separately.

### GDP over time for each country

```python
#GDP over time for each country
gdp_graphs = sns.FacetGrid(df, col='Country', col_wrap=3, hue='Country', palette='bright', sharey=False)
gdp_graphs.map(sns.lineplot, 'Year', 'GDP')
gdp_graphs.set_axis_labels('Year', 'GDP in U.S. Dollars')
gdp_graphs.set_titles('GDP - {col_name}')
gdp_graphs.add_legend()
plt.show()
```

### Life expectancy over time for each country

```python
#Life expectancy over time for each country
leaby_graphs = sns.FacetGrid(df, col='Country', col_wrap=3, hue='Country', palette='bright', sharey=False)
leaby_graphs.map(sns.lineplot, 'Year', 'LEABY')
leaby_graphs.set_axis_labels('Year', 'Life expectancy at birth (years)')
leaby_graphs.set_titles('Life expectancy - {col_name}')
leaby_graphs.add_legend()
plt.show()
```

NOTE: Each graph has a different y-axis scale. This is because the range of GDP and Life expectancy is different for each country. By scaling each country individually we can see the changes in GDP and Life expectancy for each country more clearly.

After these visualizations we can say that GDP and Life expectancy rise for each country over time.
Next we will calculate the correlation between GDP and Life expectancy for each country.

### Correlation between GDP and Life expectancy for each country
We use Linear Regression to visualize the correlation between GDP and Life expectancy for each country.

```python
corr_graphs = sns.FacetGrid(df, col='Country', col_wrap=3, hue='Country', palette='bright', sharey=False, sharex=False)
corr_graphs.map(sns.regplot, 'GDP', 'LEABY')
corr_graphs.set_axis_labels('GDP in U.S. Dollars', 'Life expectancy at birth (years)')
corr_graphs.set_titles('Correlation - {col_name}')
corr_graphs.add_legend()
plt.show()
```

### Summary

After this short analysis we can say that there is a strong linear relationship between GDP and Life expectancy for each country.
We can also say that GDP and Life expectancy rise for every country over time.
